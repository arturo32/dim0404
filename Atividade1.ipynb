```python
import numpy as np
import matplotlib.pyplot as plt

# definição de funções
def plot_xy(x, y, title):
    fig, ax = plt.subplots(1,figsize=(10,5))
    ax.plot(x,y)
    ax.set_title(title)
    ax.set_xlabel('Valor de x')
    ax.set_ylabel('valor de y')
    #spine placement data centered
    ax.spines['left'].set_position(('data', 0.0))
    ax.spines['bottom'].set_position(('data', 0.0))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    plt.show()
```

```python
#3
x1 = np.arange(-10, -1, 0.1)
x2 = np.arange(-0.9, 2, 0.1)
x3 = np.arange(2.1, 10, 0.1)
y1 = x1 + 1
y2 = -np.power(x2, 2) + 2
y3 = x3
plot_xy(np.concatenate((x1,x2, x3), axis=0), np.concatenate((y1,y2, y3), axis=0), '1')
```
